```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('../input/modified-genre/genres_v2_modified.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
dataset.genre = LabelEncoder().fit_transform(dataset.genre)
```

```python
x = dataset.iloc[:,:11].values
y = dataset.iloc[:,11].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,shuffle = True,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,[2,3,10]]=sc.fit_transform(X_train[:,[2,3,10]])
X_test[:,[2,3,10]]=sc.transform(X_test[:,[2,3,10]])

```

```python
pd.DataFrame(X_train)
```

## Random Forest

```python
# Training the Random Forest Classification model on the Training set
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(bootstrap = False,n_estimators = 1000, 
                                    criterion = 'entropy', random_state = 1,
                                    max_features='sqrt')
classifier.fit(X_train, y_train)

```

```python
y_pred = classifier.predict(X_test)
p = {'pred':y_pred,'true':y_test}
pd.DataFrame(p)
```

```python

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

## Neutral Network

```python
from sklearn.neural_network import MLPClassifier

classifier = MLPClassifier(max_iter=500, hidden_layer_sizes = (150,),
                           activation= 'relu', solver = 'adam',
                           learning_rate= 'invscaling',random_state = 0)

classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
p = {'pred':y_pred,'true':y_test}
pd.DataFrame(p)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
